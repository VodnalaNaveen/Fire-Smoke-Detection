```python
source = cv2.VideoCapture(0)

onnx_model = YOLO("New folder/Fire-Smoke-1/best.onnx") 
pygame.mixer.init()
    # Load your audio file
pygame.mixer.music.load('audio.mp3') 


while True:
    ret, frame = source.read()
    if not ret:
        break

    results = onnx_model.predict(frame)
    result_frame = results[0].plot()
    cv2.imshow('Prediction', result_frame)
    if len(results[0]) >=1 :
        if not pygame.mixer.music.get_busy():
            pygame.mixer.music.play()
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


source.release()
cv2.destroyAllWindows()
pygame.mixer.music.stop()
```

### To save each frame after detection

```python
source = cv2.VideoCapture(0)

onnx_model = YOLO("Fire-Smoke-1/best.onnx") 


output_dir = "webcam_output"
os.makedirs(output_dir, exist_ok=True)

frame_count = 0

while True:
    ret, frame = source.read()
    if not ret:
        break

    results = onnx_model.predict(frame)
    result_frame = results[0].plot()

    frame_path = os.path.join(output_dir, f"frame_{frame_count:04d}.jpg")
    cv2.imwrite(frame_path, result_frame)

    cv2.imshow('Prediction', result_frame)
    frame_count += 1

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Cleanup
source.release()
cv2.destroyAllWindows()
```
